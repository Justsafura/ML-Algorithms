# Hierarchical Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
```

```python
X = dataset.iloc[:,[3,4]].values
```

## Using Dendogram to find the number of clusters

```python
import scipy.cluster.hierarchy as sch
```

```python
dendgram = sch.dendrogram(sch.linkage(X,method='ward')) # ward is used to minimize variance within each cluster
plt.title("Dendogram")
plt.xlabel("Customers")
plt.ylabel("Euclidian Distance")
```

** The optimal number of clusters is 5 as we see from dendogram.**

## Fitting Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ahc = AgglomerativeClustering(n_clusters=5)
y_ahc = ahc.fit_predict(X)
```

## Visualizing the Clusters

```python
plt.scatter(X[(y_ahc == 0),0],X[(y_ahc == 0),1], s = 100, c="red",label="Careful",edgecolors='black')
plt.scatter(X[(y_ahc == 1),0],X[(y_ahc == 1),1], s = 100, c="blue",label="Standard",edgecolors='black')
plt.scatter(X[(y_ahc == 2),0],X[(y_ahc == 2),1], s = 100, c="green",label="Target",edgecolors='black')
plt.scatter(X[(y_ahc == 3),0],X[(y_ahc == 3),1], s = 100, c="cyan",label="Careless",edgecolors='black')
plt.scatter(X[(y_ahc == 4),0],X[(y_ahc == 4),1], s = 100, c="magenta",label="Sensible",edgecolors='black')
plt.title("Hierarchical Clustering")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
```
